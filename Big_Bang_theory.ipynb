```python
from wordcloud import WordCloud , STOPWORDS , ImageColorGenerator

```

```python
#list of the main character

person = ['Sheldon' , 'Leonard' , 'Raj' , 'Penny' , 'Howard' , 'Amy' , 'Bernadette']

person
```

```python
data1 = data[data.person_scene.isin(person)]

len(data)
```

```python
sheldon = ''.join(dialogue for dialogue in data1[data1.person_scene == 'Sheldon'].dialogue)
Leonard = ''.join(dialogue for dialogue in data1[data1.person_scene == 'Leonard'].dialogue)
Raj = ''.join(dialogue for dialogue in data1[data1.person_scene == 'Raj'].dialogue)
Howard = ''.join(dialogue for dialogue in data1[data1.person_scene == 'Howard'].dialogue)

```

```python
stop_word = set(STOPWORDS)
stop_word.update(["Leonard","Penny","Sheldon","Raj","Howard","Amy","Bernadette","Ramona","Beverly","Cooper","Dr","Leslie","Priya",
                  "one","two","three","first","second","a","an",
                  "it","s","ve","got","make","go","making","going","are","re","do","don","t","give","see","take","think","isn","want",
                  "will","ll","did","didn","let","know","say","need","haven","look","may","come","making","does","doesn","said","made","put",
                  "what","how","m","r","e","u","d","USA","India",
                  "Huh","Uh","Oh","Yes","No","Well","Really","right","Okay","Yeah","So","now","Ew","Ow",
                  "good","bad","ever","never","back","much","many","something","anything","some","any","Maybe","someone","anyone","somebody","anybody",
                  "may","might","Of","course","hello","please","sure",
                  "I","You","me","your","his","her","mine","us","we","them" , 'guy','Hey' , 'hey' , 'Thank' , 'gonna'
                 ])

len(stop_word)
```

# sheldon most common worlds

```python
wordcloud3 = WordCloud(max_words = 70 , stopwords = stop_word , background_color= 'white').generate(sheldon)


plt.imshow(wordcloud3  , interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Raj most common worlds

```python
wordcloud = WordCloud(max_words = 70 , stopwords = stop_word , background_color= 'white').generate(Raj)


plt.imshow(wordcloud  , interpolation='bilinear')
plt.axis('off')
plt.show()
```
